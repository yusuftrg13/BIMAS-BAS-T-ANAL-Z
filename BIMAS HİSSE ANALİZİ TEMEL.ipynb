```python
import pandas as pd
import matplotlib.pyplot as plt

#Tarihi çevirelim
df['Tarih']=pd.to_datetime(df['Tarih'],format='%d.%m.%Y')

#Kapanış Fiyat Trendi
df.set_index('Tarih')['Şimdi'].plot(figsize=(10,6),title='Kapanış Fiyat Trend Analizi')

plt.xlabel("Tarih")
plt.ylabel("Kapanış Fiyatı")
plt.tight_layout()
plt.show()
```

Hareketli Ortamaları Alalım

```python
#20 günlük ve 50 günlük hareketli ortalamaları alalım
df['20_Gunluk_MA']=df['Şimdi'].rolling(window=20).mean()
df['50_Gunluk_MA']=df['Şimdi'].rolling(window=20).mean()

df[['Şimdi','20_Gunluk_MA','50_Gunluk_MA']].plot(figsize=(12,6),title='20 VE 50 GÜNLÜK HAREKETLİ ORTALAMA')

plt.xlabel("Tarih")
plt.ylabel("Fiyat")
plt.grid(True)
plt.tight_layout()
plt.show()
```

20 ve 50 günlük hareketli ortalamaların yer aldığı grafik bu grafik kısa ve uzun vadeli fiyat trendlerini anlamanıza yardımcı olur

20 günlük kısa vadeli trendleri gösterirken 50 günlük hareketli ortalama uzun vadeli trendi gösterir.Hareketli ortalamalar bize trendin ne yönde olduğunu gösterir

```python
#Günlük Getiri Hesaplayıp Çalışmayı Sonlandıracağım
df['Günlük Getiri']=df['Şimdi'].pct_change()


# Getiri trendini görselleştirelim
df['Günlük Getiri'].plot(figsize=(10, 6), title="Günlük Getiri")
plt.xlabel("Tarih")
plt.ylabel("Günlük Getiri")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Fiyat ve hacim arasındaki korelasyonu hesaplayalım
correlation = df[['Şimdi', 'Hac.']].corr()
print("Kapanış Fiyatı ile Hacim Arasındaki Korelasyon:")
print(correlation)
```
